```python
# Read the inequality in income data into a dataframe
inequality_df = pd.read_csv('/content/Inequality in Income.csv')
inequality_df.head(10)
```

```python
# Change inequality data from wide to long, making year a column
inequality_df_long = pd.melt(
    inequality_df,
    id_vars=["Country", "ISO3", "Continent", "Hemisphere", "Human Development Groups", "UNDP Developing Regions", "HDI Rank (2021)"],
    var_name="year",
    value_name="Income Inequality"
)
# Remove 'Inequality in income' and parentheses from the year column
inequality_df_long["year"] = inequality_df_long["year"].str.extract('(\d{4})').astype(int)
# Drop irrelevant features to our problem from the dataframe
# We are mainly focused on income inequality (Gini index) and possibly HDI rank, seeing if there is a correlation with overall population happiness
inequality_df_long = inequality_df_long.drop(columns = ['Continent', 'Hemisphere', 'ISO3', 'Human Development Groups', 'UNDP Developing Regions'])
print(f"Missing income inequality data: {inequality_df_long['Income Inequality'].isnull().sum()} entries")
# Some countries do not disclose this data, so we remove them from the dataframe
inequality_df_long = inequality_df_long.dropna(subset = 'Income Inequality')
inequality_df_long.head(10)
```

```python
# Subset the inequality dataset to only include the year 2021 for all countries
inequality_df_long = inequality_df_long[inequality_df_long['year'] == 2021]
inequality_df_long.head()
```

```python
# Subset the happiness dataset to only include 2021 for all countries
# 2021 is the most recent year for both datasets, so we will use this year in our analysis
happiness_df = happiness_df[happiness_df['year'] == 2021]
# Make sure column names are the same before merging
happiness_df = happiness_df.rename(columns = {'Country name': 'Country'})
happiness_df.head(10)
```

```python
# Perform the merge
merged_data_2021 = pd.merge(happiness_df, inequality_df_long, on='Country', how='inner')
merged_data_2021 = merged_data_2021.drop(columns='year_y')
merged_data_2021 = merged_data_2021.rename(columns={'year_x': 'Year'})
merged_data_2021 # I changed here to  know how many conutries we used for analytics
```

```python
# Create a correlation heat map between income inequality and features from the happiness data
correlation_data = merged_data_2021[['Income Inequality', 'Life Ladder', 'Log GDP per capita', 'Social support',
                                     'Healthy life expectancy at birth', 'Freedom to make life choices',
                                     'Generosity', 'Perceptions of corruption', 'Positive affect',
                                     'Negative affect', 'HDI Rank (2021)']]
correlation_matrix = correlation_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
```

```python
# Scatterplot to visualize correlation between income inequality and healthy life expectancy at birth
import matplotlib.pyplot as plt
plt.scatter(merged_data_2021['Income Inequality'], merged_data_2021['Healthy life expectancy at birth'], color='b')
sns.regplot(x='Income Inequality', y='Healthy life expectancy at birth', data=merged_data_2021,line_kws={"color": "red"}, ci=None)
plt.title('Income Inequality vs Healthy Life Expectancy', size=15)
plt.xlabel('Income Inequality', size=12)
plt.ylabel('Healthy Life Expectancy at Birth', size=12)
```

```python
# Scatterplot to visualize correlation between income inequality and life ladder
import matplotlib.pyplot as plt
plt.scatter(merged_data_2021['Income Inequality'], merged_data_2021['Life Ladder'], color='b')
sns.regplot(x='Income Inequality', y='Life Ladder', data=merged_data_2021,line_kws={"color": "red"}, ci=None)
plt.title('Income Inequality vs Life Ladder', size=15)
plt.xlabel('Income Inequality', size=12)
plt.ylabel('Life Ladder', size=12)
```

```python
# Scatterplot to visualize correlation between income inequality and Log GDP per capita
import matplotlib.pyplot as plt
plt.scatter(merged_data_2021['Income Inequality'], merged_data_2021['Log GDP per capita'], color='b')
sns.regplot(x='Income Inequality', y='Log GDP per capita', data=merged_data_2021,line_kws={"color": "red"}, ci=None)
plt.title('Income Inequality vs Log GDP per capita', size=15)
plt.xlabel('Income Inequality', size=12)
plt.ylabel('Log GDP per capita', size=12)
```

```python
# Scatterplot to visualize correlation between income inequality and Negative affect
import matplotlib.pyplot as plt
plt.scatter(merged_data_2021['Income Inequality'], merged_data_2021['Negative affect'], color='b')
sns.regplot(x='Income Inequality', y='Negative affect', data=merged_data_2021,line_kws={"color": "red"}, ci=None)
plt.title('Income Inequality vs Negative affect', size=15)
plt.xlabel('Income Inequality', size=12)
plt.ylabel('Negative affect', size=12)
```

```python
# Remove outliers and plot again
from scipy import stats

z_scores = np.abs(stats.zscore(merged_data_2021[['Income Inequality', 'Negative affect']]))

# Define a threshold for what is considered an outlier (Z-score > 3)
threshold = 3
outliers = (z_scores > threshold)

# Remove rows where any column is an outlier
filtered_data = merged_data_2021[~outliers.any(axis=1)]

# Create the scatter plot without outliers
plt.figure(figsize=(8, 6))
sns.regplot(x='Income Inequality', y='Negative affect', data=filtered_data, scatter_kws={"color": "blue"}, line_kws={"color": "red"}, ci=None)
plt.title('Income Inequality vs Negative affect (Outliers Removed)', size=15)
plt.xlabel('Income Inequality', size=12)
plt.ylabel('Negative affect', size=12)

plt.grid(True)
plt.show()
```
